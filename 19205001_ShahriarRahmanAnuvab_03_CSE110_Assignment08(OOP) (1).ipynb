```python
#Write your code here

class Calculator:
  def __init__(self,a,b,operator):
    self.a = a
    self.b = b
    self.operator = operator

  def calculate_button(self):
    if(self.operator == '+'):
      return self.a+self.b
    elif(self.operator == '-'):
      return self.a-self.b
    elif(self.operator == '*'):
      return self.a*self.b
    elif(self.operator == '/'):
      return self.a/self.b
  def details(self):
    if(self.operator == '+'):
      print(self.a,self.operator,self.b,"=",self.calculate_button())
    elif(self.operator == '-'):
      print(self.a,self.operator,self.b,"=",self.calculate_button())
    elif(self.operator == '*'):
      print(self.a,self.operator,self.b,"=",self.calculate_button())
    elif(self.operator == '/'):
      print(self.a,self.operator,self.b,"=",self.calculate_button())

#Do not change the following code:
add = Calculator(10, 20, '+')
val = add.calculate_button()
print("Returned value:", val)
add.details()
print("==================")

sub = Calculator(val, 10, '-')
val = sub.calculate_button()
print("Returned value:", val)
sub.details()
print("==================")

div = Calculator(val, 5, '*')
val = div.calculate_button()
print("Returned value:", val)
div.details()
print("==================")

mult = Calculator(val, 16, '/')
val = mult.calculate_button()
print("Returned value:", val)
mult.details()
print("==================")
```

•	The default <strong>fare charge</strong> of any vehicle is seating capacity * 100 and add an extra 10% on full fare as a maintenance charge. So total fare for a vehicle instance will become the final amount = total fare + 10% of the total fare.<br>

<br>
COPY THIS PORTION:<br>
--------------------------------------<br>
<em>#Write your code here<br><br>
#Do not change the following code:</em><br>
School_bus = Vehicle("School Volvo", 180, 12)<br>

--------------------------------------<br>

<strong>OUTPUT:</strong><br>

Vehicle Name: School Volvo<br>
Speed: 180 <br>
Mileage: 12<br>
The seating capacity of a bus is 50 passengers<br>
Total Bus fare is: 5500.0<br>


```python
#Write your code here
class Vehicle:
  seating_cap = 50

  def __init__(self,name,speed,milage):
    self.name =name
    self.speed = speed
    self.milage = milage

  def seating_capacity(self):
    str1=""
    str1 = str1 + "Vehicle Name: " + self.name + "\nSpeed: " + str(self.speed) + "\nMilage: " + str(self.milage) + "\nThe seating capacity of a bus is " + str(self.seating_cap) + " passengers"
    return str1
  
  def fare(self):
    capacity = self.seating_cap
    f = capacity*100
    total = f+f*10
    return total

#Do not change the following code:
School_bus = Vehicle("School Volvo", 180, 12)
print(School_bus.seating_capacity())
print("Total Bus fare is:", School_bus.fare())
```

```python
#Write your code here

class Student:
  def __init__(self,name,id,dept,grade):
    self.name = name
    self.id = id
    self.dept = dept
    self.grade = grade
    print("--------- Department:",self.dept,"---------")
    print("Name:",self.name,end=',')
    print("ID:",self.id)

  def calculate_CGPA(self):
    sum = 0
    for g in self.grade:
      sum = sum+g*3
    cgpa = sum/len(self.grade*3)
    #print(cgpa)
    return cgpa

  def print_details(self):
    result = self.calculate_CGPA()
    print("CGPA:",result)
    comment = ["Highest Distinction","High Distinction","Distinction","Satisfactory"]

    if(3.80<=result<=4.00):
      print("Your Academic standing is",comment[0]) 
    elif(3.65<=result<=3.79):
      print("Your Academic standing is",comment[1])
    elif(3.50<=result<=3.64):
      print("Your Academic standing is",comment[2])
    elif(2.00<=result<=3.49):
      print("Your Academic standing is",comment[3])
    elif(0<=result<=1.99):
      print("You cannot graduate")
    

#Do not change the following code:
s1 = Student('Dora', '15995599','CSE', [4,3.7,3.7,4])
s1.calculate_CGPA()
s1.print_details()

s2 = Student('Pingu', '12312322', 'EEE', [1.7,1.3,1.3,1.3,1])
s2.calculate_CGPA()
s2.print_details()

s3 = Student('Bob', '13311331', 'CSE', [2,3,3,3.7,2.7,2.7])
s3.calculate_CGPA()
s3.print_details()
```
